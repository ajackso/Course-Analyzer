```python
from plotly.offline import plot
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
init_notebook_mode() # run to allow plots to show up in Notebook
import plotly.graph_objs as go
from plotly import tools
```

```python
def getTerm(df):
    b = df.term > 0
    blist = ["Fall" if x > 0 else "Spring" for x in b]
    return blist
```

```python
cs111Data = go.Scatter(
    x = cs111DF.year+cs111DF.term,
    y = cs111DF.enrollment,
    mode = 'markers+lines',
    marker=dict(
        size='16',
        color = 'rgb(67,129,179)',
#         colorscale='Blues',
#         showscale=False
        line = dict(
            width = 1,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = "Prof. " + cs111DF.instructor + "<br>" + getTerm(cs111DF),
    textposition='left',
    line=dict(
        shape='vhv' # or linear
    )
)
layout = go.Layout(          
    title = "CS 111 course enrollments, 2010-2016",  
    xaxis = dict(                 
        title="Year"        
       ),
    yaxis = dict(                 
        title="Course enrollment"        
       )
    )
    
fig = go.Figure(data=[cs111Data], layout=layout)

iplot(fig)
```

```python
yearList = cs111DF.year.tolist()
yearListStr = [str(yr) for yr in yearList]
print yearListStr
```

```python
cs111Data = go.Bar(
    x = cs111DF.year.sum(),
    y = cs111DF.enrollment,
    text = "Prof. " + cs111DF.instructor + "<br>" + getTerm(cs111DF)+ " " + yearListStr,
    name = 'enrollment',
    marker=dict(
        color='rgb(121,186,233)'
    )
) #(15,117,188)
cs111Max = go.Scatter(
        x = cs111DF.year.sum(),
        y=cs111DF.max_enrollment,
        name = "max enrollment",
        marker=dict(
        color='rgb(15,117,188)'
    )
        ) 
layout = go.Layout(          
    title = "CS 111 course enrollments, 2010-2016",  
    xaxis = dict( 
        title="Year",
        ticktext = yearListStr,
        tickvals = range(len(yearListStr)),
        tickangle=-45
       ),
    yaxis = dict(                 
        title="Course enrollment"        
       )
#     bargap=0.1
    )
    
fig = go.Figure(data=[cs111Data,cs111Max], layout=layout)

iplot(fig)
```
